```python
# LIbrary Importing
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from sklearn.linear_model import LinearRegression
import pandas as pd
import os
import numpy as np
```

```python
# Variable Assigning
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "apex", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]

YEAR = 2025
all_data = []

TIME_LIMIT = 300
```

```python
# Fetching Algorithm
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    wait = WebDriverWait(driver, 10)
    try:
        for team in ufa_teams:
            print(f"Processing team: {team}, year: {YEAR}")
            url = f"https://watchufa.com/stats/player-stats?team={team}&year={YEAR}"
            driver.get(url)
            try:
                table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                table_html = table.get_attribute("outerHTML")
                tables = pd.read_html(table_html)
                for t in tables:
                    t["Team"] = team
                    t["Year"] = YEAR
                    all_data.append(t)
            except Exception as e:
                print(f"Failed to get data for {team} {YEAR}: {e}")
    finally:
        driver.quit()
scrape_all()
```

```python
# Combining and editing
players = pd.concat(all_data, ignore_index=True)

players = players.rename(columns={'+/- â–¼':'+/-'})
```

```python
# Standardizing
stats_col = {
    'SCR': 'SCR_Z',
    'BLK': 'BLK_Z',
    '+/-': 'PM_Z',
    'Y': 'Y_Z',
    'HA': 'HA_Z',
    'Cmp': 'CMP%_Z',
    'AST':'AST_Z',
    'GLS':'GLS_Z',
    'Cmp':'CMP_Z',
    'T':'T_Z',
    'RY': 'RY_Z',
    'TY':'TY_Z',
    'Y':'Y/CMP_Z',
    'OEFF':'OEFF_Z'
}

mean = players[list(stats_col.keys())].mean()
std = players[list(stats_col.keys())].std()

for col, z_col in stats_col.items():
    players[z_col] = ((players[col] - mean[col]) / std[col]).round(2)

players['team_avg_min'] = (
    players.groupby('Team')['MP']
           .transform('mean')
           .round()
)
```

```python
# Linear Regression Offensive
OFF_I = players[['HA_Z','CMP_Z','GLS_Z','AST_Z','T_Z','TY_Z']]
OFF_O = players['OEFF_Z']

model = LinearRegression(fit_intercept=True)
model.fit(OFF_I, OFF_O)

weights = pd.Series(
    model.coef_,
    index=OFF_I.columns
)
intercept = model.intercept_
print(weights)
```

```python
# Calculating Scores
players['Offensive_Value'] = round(OFF_I.dot(weights) + intercept,2)
players['Score'] = (players['Offensive_Value'] * np.sqrt(players['MP'] / players['team_avg_min'])).round(3)
```

```python
# Saving to Excel
players.to_csv("2 Weights Non Per Possession.csv", index=False)
os.startfile("2 Weights Non Per Possession.csv")
```
