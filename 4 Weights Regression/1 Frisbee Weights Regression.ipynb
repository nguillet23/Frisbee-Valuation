```python
# LIbrary Importing
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from sklearn.linear_model import LinearRegression
import pandas as pd
import os
import numpy as np
```

```python
# Variable Assigning
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "apex", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]

YEAR = 2025
all_data = []

TIME_LIMIT = 300
```

```python
# Fetching Algorithm
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    wait = WebDriverWait(driver, 10)
    try:
        for team in ufa_teams:
            print(f"Processing team: {team}, year: {YEAR}")
            url = f"https://watchufa.com/stats/player-stats?team={team}&year={YEAR}"
            driver.get(url)
            try:
                table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                table_html = table.get_attribute("outerHTML")
                tables = pd.read_html(table_html)
                for t in tables:
                    t["Team"] = team
                    t["Year"] = YEAR
                    all_data.append(t)
            except Exception as e:
                print(f"Failed to get data for {team} {YEAR}: {e}")
    finally:
        driver.quit()
scrape_all()
```

```python
# Combining and editing
players = pd.concat(all_data, ignore_index=True)

players = players.rename(columns={'+/- â–¼':'+/-'})
players['AST/POS %'] = round(players['AST']/players['POS'] * 100, 2)
players['GLS/POS %'] = round(players['GLS']/players['POS'] * 100, 2)
players['SCR/POS %'] = round(players['SCR'] / players['POS'] * 100, 2)
players['CMP/POS'] = round(players['Cmp'] / players['POS'], 2)
players['BLK/PP %'] = round(players['BLK']/players['PP'] * 100, 2)
players['PM/PP'] = round(players['+/-']/players['PP'], 2)
players['TY/POS'] = round(players['TY']/players['POS'])
players['RY/POS'] = round(players['RY']/players['POS'])
players['Y/POS'] = round(players['Y']/players['POS'])
players['HA/POS %'] = round(players['HA']/players['POS'] * 100, 2)
players['T/POS %'] = round(players['T']/players['POS'] * 100, 2)
players['Y/CMP'] = round(players['Y']/players['Cmp'] * 100, 2)
```

```python
# Standardizing
stats_col = {
    'SCR/POS %': 'SCR_Z',
    'BLK/PP %': 'BLK_Z',
    'PM/PP': 'PM_Z',
    'Y/POS': 'Y_Z',
    'HA/POS %': 'HA_Z',
    'Cmp%': 'CMP%_Z',
    'AST/POS %':'AST_Z',
    'GLS/POS %':'GLS_Z',
    'CMP/POS':'CMP_Z',
    'T/POS %':'T_Z',
    'RY/POS': 'RY_Z',
    'TY/POS':'TY_Z',
    'Y/CMP':'Y/CMP_Z',
    'OEFF':'OEFF_Z'
}

mean = players[list(stats_col.keys())].mean()
std = players[list(stats_col.keys())].std()

for col, z_col in stats_col.items():
    players[z_col] = ((players[col] - mean[col]) / std[col]).round(2)

players['team_avg_min'] = (
    players.groupby('Team')['MP']
           .transform('mean')
           .round()
)
```

```python
# Linear Regression Offensive
OFF_I = players[['HA_Z','CMP_Z','GLS_Z','AST_Z','T_Z','TY_Z']]
OFF_O = players['OEFF_Z']

model = LinearRegression(fit_intercept=True)
model.fit(OFF_I, OFF_O)

weights = pd.Series(
    model.coef_,
    index=OFF_I.columns
)
intercept = model.intercept_
print(weights)
```

```python
players['Offensive_Value'] = round(OFF_I.dot(weights) + intercept,2)
players['Final Score'] = (players['Offensive_Value'] * np.sqrt(players['MP'] / players['team_avg_min'])).round(3)
```

```python
# Building Subscores
players['OFF_VALUE'] = round(.3 * players['SCR_Z'] + .25 * players['HA_Z'] + .25 * players['Y_Z'] + .2 * players['CMP%_Z'],3)
players['Value'] = round(.75 * players['OFF_VALUE'] + .25 * players['PM_Z'],3)
players['Score'] = (players['Value'] * np.sqrt(players['MP'] / players['team_avg_min'])).round(3)
```

```python
# Saving to Excel
players.to_csv("1 Weights Regression.csv", index=False)
os.startfile("1 Weights Regression.csv")
```
