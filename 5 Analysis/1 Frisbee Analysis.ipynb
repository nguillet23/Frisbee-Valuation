```python
# LIbrary Importing
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from sklearn.linear_model import LinearRegression
import pandas as pd
import os
import numpy as np
from itertools import combinations
import matplotlib.pyplot as plt
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
# Variable Assigning
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "apex", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]

YEAR = 2023
all_data = []

TIME_LIMIT = 300
```

```python
# Fetching Algorithm
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options )
    wait = WebDriverWait(driver, 10)
    try:
        for team in ufa_teams:
            print(f"Processing team: {team}, year: {YEAR}")
            url = f"https://watchufa.com/stats/player-stats?team={team}&year={YEAR}"
            driver.get(url)
            try:
                table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                table_html = table.get_attribute("outerHTML")
                tables = pd.read_html(table_html)
                for t in tables:
                    t["Team"] = team
                    t["Year"] = YEAR
                    all_data.append(t)
            except Exception as e:
                print(f"Failed to get data for {team} {YEAR}: {e}")
    finally:
        driver.quit()
scrape_all()
```

```python
# Combining and editing
players = pd.concat(all_data, ignore_index=True)

players = players.rename(columns={'+/- â–¼':'+/-'})
players['AST/POS %'] = round(players['AST']/players['POS'] * 100, 2)
players['GLS/POS %'] = round(players['GLS']/players['POS'] * 100, 2)
players['SCR/POS %'] = round(players['SCR'] / players['POS'] * 100, 2)
players['CMP/POS'] = round(players['Cmp'] / players['POS'], 2)
players['BLK/PP %'] = round(players['BLK']/players['PP'] * 100, 2)
players['PM/PP'] = round(players['+/-']/players['PP'], 2)
players['TY/POS'] = round(players['TY']/players['POS'])
players['RY/POS'] = round(players['RY']/players['POS'])
players['Y/POS'] = round(players['Y']/players['POS'])
players['HA/POS %'] = round(players['HA']/players['POS'] * 100, 2)
players['T/POS %'] = round(players['T']/players['POS'] * 100, 2)
players['Y/CMP'] = round(players['Y']/players['Cmp'], 2)
```

```python
# League and Team Average Stats
players['team_avg_min'] = (
    players.groupby('Team')['MP']
           .transform('mean')
           .round()
)
league_avg_SCR = round(players['SCR/POS %'].mean())
league_avg_BLK = round(players['BLK/PP %'].mean())
league_avg_PM = round(players['PM/PP'].mean())
league_avg_Y = round(players['Y/POS'].mean())
league_avg_HA = round(players['HA/POS %'].mean())
league_avg_CMP = round(players['CMP/POS'].mean())
league_avg_AST = round(players['AST/POS %'].mean())
league_avg_GLS = round(players['GLS/POS %'].mean())
league_avg_T = round(players['T/POS %'].mean())
league_avg_TY = round(players['TY/POS'].mean())
league_avg_RY = round(players['RY/POS'].mean())
league_avg_YPerCMP = round(players['Y/CMP'].mean())
league_avg_MP = round(players['MP'].mean())
```

```python
# Shrinking
k = 5
league_avg = {
    'SCR': league_avg_SCR,
    'BLK': league_avg_BLK,
    '+/-': league_avg_PM,
    'Y': league_avg_Y,
    'HA': league_avg_HA,
    'Cmp': league_avg_CMP,
    'AST': league_avg_AST,
    'GLS': league_avg_GLS,
    'T': league_avg_T,
    'TY': league_avg_TY,
    'RY': league_avg_RY,
    'Y/CMP':league_avg_YPerCMP,
    'MP':league_avg_MP
}
for stat, avg in league_avg.items():
     players[stat + '_adj'] = round((players[stat] + k * avg) / (players['POS'] + k),2)
```

```python
# Standardizing
stats_col = {
    'SCR_adj': 'SCR_Z',
    'BLK_adj': 'BLK_Z',
    '+/-_adj': 'PM_Z',
    'Y_adj': 'Y_Z',
    'HA_adj': 'HA_Z',
    'Cmp_adj':'CMP_Z',
    'AST_adj':'AST_Z',
    'GLS_adj':'GLS_Z',
    'T_adj':'T_Z',
    'RY_adj': 'RY_Z',
    'TY_adj':'TY_Z',
    'OEFF':'OEFF_Z',
    'Y/CMP':'YperCMP_Z',
    'MP_adj':'MP_z'
}

mean = players[list(stats_col.keys())].mean()
std = players[list(stats_col.keys())].std()

for col, z_col in stats_col.items():
    players[z_col] = ((players[col] - mean[col]) / std[col]).round(2)
```

```python
# Linear Regression Offensive
OFF_I = players[['HA_Z','CMP_Z','GLS_Z','AST_Z','T_Z','TY_Z','RY_Z']]
OFF_O = players['OEFF_Z']

model = LinearRegression(fit_intercept=True)
model.fit(OFF_I, OFF_O)

weights = pd.Series(
    model.coef_,
    index=OFF_I.columns
)
intercept = model.intercept_
print(weights)
```

```python
# Calculating Scores
players['Offensive_Value'] = round(OFF_I.dot(weights) + intercept,2)
players['Score'] = (players['Offensive_Value'] * np.sqrt(players['MP'] / players['team_avg_min'])).round(3)
```

```python
# Team Average Score
players['team_avg_scr'] = (
    players.groupby('Team')['Score']
           .transform('mean')
           .round(3)
)
```

```python
# Making team score average bar chart
team_scores = (
    players
    .groupby('Team')['team_avg_scr']
    .first()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
team_scores.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Average Score')
plt.title('Team Average Scores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Saving Complete data to Excel
players.to_csv("1 Frisbee Analysis.csv", index=False)
os.startfile("1 Frisbee Analysis.csv")
```

Start Analysis of Experiment

```python
# Separating players into different 'tiers'
analysis = players.loc[:, ~players.columns.str.contains('z', case=False)]
analysis = analysis.loc[:, ~analysis.columns.str.contains('adj', case=False)]
best = analysis[analysis['Score'] >= 1].copy().sort_values('Score')
solid = analysis[(analysis['Score'] >= 0.75) & (analysis['Score'] < 1)].copy().sort_values('Score')
mediocre = analysis[(analysis['Score'] >= 0.5) & (analysis['Score'] < .75)].copy().sort_values('Score')
bad = analysis[(analysis['Score'] >= 0) & (analysis['Score'] < 0.5)].copy().sort_values('Score')
worst = analysis[(analysis['Score'] <= 0)].copy().sort_values('Score')
```

Creating Charts to Compare Players in each tier

```python
plt.figure(figsize=(10, 6))
plt.barh(best['Player'], best['Score'])
plt.xlabel('Score')
plt.ylabel('Player')
plt.title('Best Players')
plt.show()
```

```python
# Creating one group that has the averages of each tier, and the best players
best = best.sort_values(by='Score',ascending=False).reset_index(drop=True)
groups = {'worst_avg': worst, 'bad_avg': bad, 'mediocre_avg': mediocre, 'solid_avg': solid, 'best_average': best}
avg_dfs = []

for name, df in groups.items():
    avg_df = pd.DataFrame([df.mean(numeric_only=True).round(2)])
    avg_df.insert(0, 'Player', name)  # Name column filled
    avg_dfs.append(avg_df)

all_players = pd.concat([best] + avg_dfs, ignore_index=True)
all_players.to_csv("1 Frisbee Combined Averages.csv", index=False)
os.startfile("1 Frisbee Combined Averages.csv")
```
