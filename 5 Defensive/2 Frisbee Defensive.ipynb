```python
# Library Importing
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from sklearn.linear_model import LinearRegression
import pandas as pd
import os
import numpy as np
from itertools import combinations
import matplotlib.pyplot as plt
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import statsmodels.api as sm
```

```python
# Variable Assigning
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "apex", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]

YEAR = 2023
all_data = []

TIME_LIMIT = 300
```

```python
# Fetching Algorithm
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options )
    wait = WebDriverWait(driver, 10)
    try:
        for team in ufa_teams:
            print(f"Processing team: {team}, year: {YEAR}")
            url = f"https://watchufa.com/stats/player-stats?team={team}&year={YEAR}"
            driver.get(url)
            try:
                table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                table_html = table.get_attribute("outerHTML")
                tables = pd.read_html(table_html)
                for t in tables:
                    t["Team"] = team
                    t["Year"] = YEAR
                    all_data.append(t)
            except Exception as e:
                print(f"Failed to get data for {team} {YEAR}: {e}")
    finally:
        driver.quit()
scrape_all()
```

```python
# Combining and editing (League and Team Average Stats)
players = pd.concat(all_data, ignore_index=True)
players = players[players['G'] > 5]

players['team_avg_min'] = (players.groupby('Team')['MP'].transform('mean').round())
players['team_avg_pull'] = (players.groupby('Team')['Pul'].transform('mean').round())
players['team_avg_blk'] = (players.groupby('Team')['BLK'].transform('mean').round())

players = players.rename(columns={'+/- ▼':'+/-'})
players['AST/POS %'] = round(players['AST']/players['POS'] * 100, 2)
players['GLS/POS %'] = round(players['GLS']/players['POS'] * 100, 2)
players['SCR/POS %'] = round(players['SCR'] / players['POS'] * 100, 2)
players['CMP/POS'] = round(players['Cmp'] / players['POS'], 2)
players['PM/PP'] = round(players['+/-']/players['PP'], 2)
players['TY/POS'] = round(players['TY']/players['POS'])
players['RY/POS'] = round(players['RY']/players['POS'])
players['Y/POS'] = round(players['Y']/players['POS'])
players['HA/POS %'] = round(players['HA']/players['POS'] * 100, 2)
players['T/POS %'] = round(players['T']/players['POS'] * 100, 2)
players['Y/CMP'] = round(players['Y']/players['Cmp'], 2)
players['BLK/PP %'] = round(players['BLK']/players['PP'] * 100, 2)
players['Pull/Team Pull %'] = round(players['Pul']/players['team_avg_pull'] * 100, 2)
players['BLK/Team avg BLK %'] = round(players['BLK']/players['team_avg_blk'] * 100, 2)

league_avg_SCR = round(players['SCR/POS %'].mean())
league_avg_BLK = round(players['BLK/PP %'].mean())
league_avg_PM = round(players['PM/PP'].mean())
league_avg_Y = round(players['Y/POS'].mean())
league_avg_HA = round(players['HA/POS %'].mean())
league_avg_CMP = round(players['CMP/POS'].mean())
league_avg_AST = round(players['AST/POS %'].mean())
league_avg_GLS = round(players['GLS/POS %'].mean())
league_avg_T = round(players['T/POS %'].mean())
league_avg_TY = round(players['TY/POS'].mean())
league_avg_RY = round(players['RY/POS'].mean())
league_avg_YPerCMP = round(players['Y/CMP'].mean())
league_avg_MP = round(players['MP'].mean())
league_avg_DPP = round(players['DPP'].mean())
league_avg_Pul = round(players['Pull/Team Pull %'].mean())
league_avg_huck = round(players['Hck'].mean())
```

```python
# Shrinking
k = 10
league_avg = {
    'SCR/POS %': league_avg_SCR,
    'BLK/PP %': league_avg_BLK,
    'PM/PP': league_avg_PM,
    'Y/POS': league_avg_Y,
    'HA/POS %': league_avg_HA,
    'CMP/POS': league_avg_CMP,
    'AST/POS %': league_avg_AST,
    'GLS/POS %': league_avg_GLS,
    'T/POS %': league_avg_T,
    'TY/POS': league_avg_TY,
    'RY/POS': league_avg_RY,
    'Y/CMP':league_avg_YPerCMP,
    'MP':league_avg_MP,
    'DPP':league_avg_DPP,
    'Pull/Team Pull %':league_avg_Pul,
    'Hck':league_avg_huck
}
for stat, avg in league_avg.items():
    players[stat + '_adj'] = round( (players[stat] * players['POS'] / (players['POS'] + k)) + (k * avg / (players['POS'] + k)), 2 )
```

```python
# Standardizing
stats_col = {
    'SCR': 'SCR_Z',
    'BLK': 'BLK_Z',
    '+/-': 'PM_Z',
    'Y': 'Y_Z',
    'HA': 'HA_Z',
    'Cmp':'CMP_Z',
    'AST':'AST_Z',
    'GLS':'GLS_Z',
    'T':'T_Z',
    'RY': 'RY_Z',
    'TY':'TY_Z',
    'OEFF':'OEFF_Z',
    'Y/CMP':'YperCMP_Z',
    'MP':'MP_Z',
    'DPP':'DPP_Z',
    'Pull/Team Pull %':'Pull_Z'
}

mean = players[list(stats_col.keys())].mean()
std = players[list(stats_col.keys())].std()

for col, z_col in stats_col.items():
    players[z_col] = ((players[col] - mean[col]) / std[col]).round(2)
```

```python
# Linear Regression Offensive
OFF_I = players[['HA_Z','SCR_Z','T_Z','TY_Z','MP_Z']]
OFF_O = players['OEFF_Z']

model = LinearRegression(fit_intercept=True)
model.fit(OFF_I, OFF_O)

weights_off = pd.Series(
    model.coef_,
    index=OFF_I.columns
)
intercept_off = model.intercept_
print(weights_off)

players['Offensive_Value'] = round(OFF_I.dot(weights_off) + intercept_off,2)
```

```python
# Linear Regression Defensive
DEF_I = players[['BLK_Z','Pull_Z']]
DEF_O = players['DPP_Z']

model = LinearRegression(fit_intercept=True)
model.fit(DEF_I, DEF_O)

weights_def = pd.Series(
    model.coef_,
    index=DEF_I.columns
)
intercept_def = model.intercept_
print(weights_def)

players['Defensive_Value'] = round(DEF_I.dot(weights_def) + intercept_def,2)
```

```python
# Linear Regression Overall
value_I = players[['Offensive_Value','Defensive_Value']]
value_O = players['PM_Z']

model = LinearRegression(fit_intercept=True)
model.fit(value_I, value_O)

weights = pd.Series(
    model.coef_,
    index=value_I.columns
)
intercept = model.intercept_
print(weights)

players['Score'] = round(value_I.dot(weights) + intercept,2)
```

```python
# Team Average Score
players['team_avg_scr'] = (players.groupby('Team')['Score'].transform('mean').round(3))
```

```python
# Separating players into different 'tiers'
analysis = players
best = analysis[analysis['Score'] >= 1.5].copy().sort_values('Score').sort_values('Score')
solid = analysis[(analysis['Score'] >= 0.75) & (analysis['Score'] < 1.5)].copy().sort_values('Score')
mediocre = analysis[(analysis['Score'] >= 0.5) & (analysis['Score'] < .75)].copy().sort_values('Score')
bad = analysis[(analysis['Score'] >= 0) & (analysis['Score'] < 0.5)].copy().sort_values('Score')
worst = analysis[(analysis['Score'] <= 0)].copy().sort_values('Score')
```

```python
# Creating one group that has the averages of each tier, and the best players
best = best.sort_values(by='Score',ascending=False).reset_index(drop=True)
groups = {'worst_avg': worst, 'bad_avg': bad, 'mediocre_avg': mediocre, 'solid_avg': solid, 'best_average': best}
avg_dfs = []

for name, df in groups.items():
    avg_df = pd.DataFrame([df.mean(numeric_only=True).round(2)])
    avg_df.insert(0, 'Player', name)  # Name column filled
    avg_dfs.append(avg_df)

all_players = pd.concat([best] + avg_dfs, ignore_index=True)
all_players.to_csv("2 Frisbee Defensive Averages.csv", index=False)
os.startfile("2 Frisbee Defensive Averages.csv")
players.to_csv("2 Frisbee Defense.csv", index=False)
os.startfile("2 Frisbee Defense.csv")
```
