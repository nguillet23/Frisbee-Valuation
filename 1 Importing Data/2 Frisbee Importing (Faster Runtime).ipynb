```python
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "summit", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]

all_data = []

TIME_LIMIT = 300
```

```python
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    wait = WebDriverWait(driver, 10)
    try:
        for year in range(2021, 2025):
            for team in ufa_teams:
                print(f"Processing team: {team}, year: {year}")
                url = f"https://watchufa.com/stats/player-stats?team={team}&year={year}"
                driver.get(url)
                try:
                    table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                    table_html = table.get_attribute("outerHTML")
                    tables = pd.read_html(table_html)
                    for t in tables:
                        t["Team"] = team
                        t["Year"] = year
                        all_data.append(t)
                except Exception as e:
                    print(f"Failed to get data for {team} {year}: {e}")
    finally:
        driver.quit()
```

```python
# Start the scraping in a separate thread
scrape_thread = threading.Thread(target=scrape_all)
scrape_thread.start()
scrape_thread.join(timeout=TIME_LIMIT)
```

```python
# Combine the data if any was collected
combined_df = pd.concat(all_data, ignore_index=True)
# Save or process your DataFrame as needed
combined_df.to_csv("Ufa_stats.csv", index=False)
```
