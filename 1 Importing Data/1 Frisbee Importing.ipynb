```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
```

```python
# List of all UFA teams (as of 2025)
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "summit", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", "bighorns"
]
all_data = []
```

```python
# Setup headless Chrome
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

for year in range(2021, 2025):
    for team in ufa_teams:
        print(f"Processing team: {team}, year: {year}")
        url = f"https://watchufa.com/stats/player-stats?team={team}&year={year}"
        driver.get(url)
        time.sleep(5)  # Wait for page to load

        try:
            table_elements = driver.find_elements("css selector", "table")
            for i, table in enumerate(table_elements):
                table_html = table.get_attribute("outerHTML")
                tables = pd.read_html(table_html)
                for t in tables:
                    t["Table"] = f"table_{i}"  # optional label
                    t["Team"] = team
                    t["Year"] = year
                    all_data.append(t)
        except Exception as e:
            print(f"Failed to get data for {team} {year}: {e}")
            
driver.quit()
```

```python
# Combine all years and teams into a single DataFrame
combined_df = pd.concat(all_data, ignore_index=True)
```

```python
# Save as CSV
combined_df.to_csv("Frisbee_stats.csv", index=False)
```
