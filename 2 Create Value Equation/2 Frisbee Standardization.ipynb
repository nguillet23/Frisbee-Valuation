```python
import threading
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# List of all UFA teams (as of 2025)
ufa_teams = [
    "hustle", "sol", "glory", "flyers", "union", "summit", "breeze", "mechanix",
    "havoc", "alleycats", "aviators", "radicals", "windchill", "royal", "empire",
    "spiders", "phoenix", "thunderbirds", "steel", "shred", "growlers", "cascades",
    "rush", 
    #"bighorns"
]
all_data = []

# Set your time limit in seconds (e.g., 300 seconds = 5 minutes)
TIME_LIMIT = 300
```

```python
def scrape_all():
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    wait = WebDriverWait(driver, 10)
    try:
        for year in range(2021, 2025):
            for team in ufa_teams:
                print(f"Processing team: {team}, year: {year}")
                url = f"https://watchufa.com/stats/player-stats?team={team}&year={year}"
                driver.get(url)
                try:
                    table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))
                    table_html = table.get_attribute("outerHTML")
                    tables = pd.read_html(table_html)
                    for t in tables:
                        t["Team"] = team
                        t["Year"] = year
                        all_data.append(t)
                except Exception as e:
                    print(f"Failed to get data for {team} {year}: {e}")
    finally:
        driver.quit()
```

```python
# Start the scraping in a separate thread
scrape_thread = threading.Thread(target=scrape_all)
scrape_thread.start()
scrape_thread.join(timeout=TIME_LIMIT)
# Combine all years and teams into a single DataFrame
combined_df = pd.concat(all_data, ignore_index=True)
combined_df.columns.values[8] = "PlusMinus"
```

```python
# Select columns to standardize
stats_to_standardize = ["Y", "SCR", "HA", "PlusMinus", "BLK"]

# Standardize the columns
scaler = StandardScaler()
standardized = scaler.fit_transform(combined_df[stats_to_standardize])
standardized_df = pd.DataFrame(standardized, columns=[f"{col}_z" for col in stats_to_standardize])

# Add standardized columns to the dataframe
for col in standardized_df.columns:
    combined_df[col] = standardized_df[col]
```

```python
# "Moneyball" Equation

combined_df["eUPV"] = (
    round(
        (
            (
                0.1 * combined_df["Y_z"] +
                0.4 * combined_df["SCR_z"] + 
                0.4 * combined_df["HA_z"] +
                1.0 * combined_df["PlusMinus_z"] +
                0.2 * combined_df["BLK_z"]
            )
            * combined_df["Cmp%"] ** 3
        ) * .0001
    )
)
```

```python
# Keep ONLY essential columns used in calculation + identifiers
essential_columns = [
    combined_df.columns[0],  # Player name (first column)
    'Y',
    'SCR',
    'HA',
    'PlusMinus',
    'BLK',
    'Cmp%',
    'eUPV'
]
combined_df = combined_df[essential_columns]

# Rename first column to 'Player' for clarity
combined_df.rename(columns={combined_df.columns[0]: 'Player'}, inplace=True)

# Sort by eUPV descending
combined_df = combined_df.sort_values(by="eUPV", ascending=False)
```

```python
# Save as CSV
combined_df.to_csv("Frisbee_Standardization.csv", index=False)
```
